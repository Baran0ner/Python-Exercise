```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,60,70,80]
```

```python
kiloListesi =[20,30,40,50,55,66,77,88]
```

```python
numpyYasListesi = np.array(yasListesi)
```

```python
numpyKilolistesi =np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKilolistesi,"g")
```

```python
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilonun yaşa göre değişimi")
plt.show()
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 =numpyDizisi1**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.show()
```

```python
benimFigure=plt.figure()
figureAxes=benimFigure.add_axes([0.4,0.4,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X EKSENİ")
figureAxes.set_ylabel("Y EKSENİ")
figureAxes.set_title("Grafik Başlığı")
plt.show()
```

```python
figure2 = plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi2,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")

plt.show()
```

```python
plt.subplots()
plt.show()
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
type(benimEksenler)
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("x Ekseni ")
plt.show()
```

```python
yeniFigure = plt.figure()


yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2, label = "numpyDizisi**2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3, label= "numpyDizisi**3")
yeniEksen.legend(loc=0,)

plt.show()
```
